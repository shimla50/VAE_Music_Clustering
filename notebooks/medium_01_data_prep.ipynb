```python
import pandas as pd
from pathlib import Path

# File paths (relative to music_vae_project)
input_path = Path("../data/raw/Music.csv")
output_path = Path("../data/subset/subset_5000_dedup.csv")

# Load CSV
df = pd.read_csv(input_path)

# Keep required columns
df = df[["spotify_id", "name", "artist", "preview"]]

# Drop rows with missing values
df = df.dropna(subset=["spotify_id", "name", "artist", "preview"])

# Remove duplicate spotify_id
df = df.drop_duplicates(subset=["spotify_id"])

# Randomly select 5000 samples
df_5000 = df.sample(n=5000, random_state=42) if len(df) >= 5000 else df

# Save output
output_path.parent.mkdir(parents=True, exist_ok=True)
df_5000.to_csv(output_path, index=False)

print("✅ File saved at:", output_path)
print("Rows:", len(df_5000))
print("Unique spotify_id:", df_5000["spotify_id"].nunique())
```
